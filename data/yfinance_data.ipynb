```python
import pandas as pd
import yfinance as yf

ticker_symbol = "BTC-USD"

# Define start and end dates
start_date = "2015-01-01"
end_date = "2024-01-31"

# Fetch historical data
BTC_data = yf.Ticker(ticker_symbol)

# Retrieve historical market data for the specified period
BTC_historical_data = BTC_data.history(start=start_date, end=end_date)
```

```python
BTC_historical_data.reset_index(inplace=True)
BTC_historical_data.head()
```

```python
BTC_historical_data['Date'] = pd.to_datetime(BTC_historical_data['Date'], unit = 'ms')
BTC_historical_data['Date'] = BTC_historical_data['Date'].dt.date
```

```python
BTC_historical_data = BTC_historical_data.drop(columns=['Dividends', 'Stock Splits'])
BTC_historical_data.head()
```

```python
ticker_symbol = "^IXIC"

start_date = "2015-01-01"
end_date = "2024-01-31"

NASDAQ_data = yf.Ticker(ticker_symbol)

NASDAQ_historical_data = NASDAQ_data.history(start=start_date, end=end_date)
```

```python
BTC = BTC_historical_data
DowJones = DJI_historical_data
SP500 = SP500_historical_data
NASDAQ = NASDAQ_historical_data
```

```python
BTC.rename(columns={
    'Open': 'BTC_Open',
    'High': 'BTC_High',
    'Low': 'BTC_Low',
    'Close': 'BTC_Close',
    'Volume': 'BTC_Volume'
}, inplace=True)

BTC.head()
```

```python
DowJones.rename(columns={
    'Open': 'DJ_Open',
    'High': 'DJ_High',
    'Low': 'DJ_Low',
    'Close': 'DJ_Close',
    'Volume': 'DJ_Volume'
}, inplace=True)

DowJones.head()
```

```python
SP500.rename(columns={
    'Open': 'SP_Open',
    'High': 'SP_High',
    'Low': 'SP_Low',
    'Close': 'SP_Close',
    'Volume': 'SP_Volume'
}, inplace=True)

SP500.head()
```

```python
NASDAQ.rename(columns={
    'Open': 'NDQ_Open',
    'High': 'NDQ_High',
    'Low': 'NDQ_Low',
    'Close': 'NDQ_Close',
    'Volume': 'NDQ_Volume'
}, inplace=True)

NASDAQ.head()
```

```python
database = BTC.merge(DowJones, on='Date', how='left')
database = database.merge(SP500, on='Date', how='left')
database = database.merge(NASDAQ, on='Date', how='left')

database.head()
```

```python
database['BTC_Shifted'] = database['BTC_Close'].shift(-1)
database.head()
```

```python
database.columns
```

```python
database.info()
```

```python
database.isnull().sum()
```

```python
database['Date'] = pd.to_datetime(database['Date'])
```

```python
from scipy.interpolate import interp1d

# Cubic spline interpolation to handle NaN values

data_interpolated = database.copy()

# Interpolate each column separately
for column in database.columns:
    # Get non-NaN values and corresponding indices
    valid_indices = database[column].notna()
    valid_data = database[column][valid_indices]
    valid_dates = database.index[valid_indices]
    
    # Perform cubic spline interpolation
    interpolator = interp1d(valid_dates, valid_data, kind='cubic', bounds_error=False)
    data_interpolated[column] = interpolator(database.index)
```

```python
data_interpolated.dropna(inplace=True)
```

```python
data_interpolated['Date'] = pd.to_datetime(data_interpolated['Date'])
```

```python
# Assuming your datetime column is named 'Date'
# Convert the 'Date' column to string
data_interpolated['Date'] = data_interpolated['Date'].astype(str)

# Extract only the date part (YYYY-MM-DD)
data_interpolated['Date'] = data_interpolated['Date'].str.split(' ').str[0]
```
