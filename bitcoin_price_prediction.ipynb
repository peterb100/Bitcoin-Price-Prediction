I added another column, named BTC_Shifted. As known, in time-series forecasting, it's common to create shifted or lagged versions of the target variable (in this case, Bitcoin prices) to use as features for prediction. The purpose of creating lagged versions of the target variable is to incorporate historical values of the target variable into the feature set, allowing the model to learn patterns and relationships between past and present values.

```python
database['BTC_Shifted'] = database['BTC_Close'].shift(-1)
database.head()
```

Since I have added data for three main indices, it is espected to have NaN values in the dataset after merging. The NaN values are the direct result of missing market activity for indices on weekends or other holidays. The Cryptomarket has uniterrupted activity and we have data for every day of our choosen period.
To handle this situation, I decided to use Cubic Splin Interpolation to fill the missing gaps. This is a good approach, based on this projects's purpose, especially with a lot of NaN values present.

```python
from scipy.interpolate import interp1d

# Cubic spline interpolation to handle NaN values

data_interpolated = database.copy()

# Interpolate each column separately
for column in database.columns:
    # Get non-NaN values and corresponding indices
    valid_indices = database[column].notna()
    valid_data = database[column][valid_indices]
    valid_dates = database.index[valid_indices]
    
    # Perform cubic spline interpolation
    interpolator = interp1d(valid_dates, valid_data, kind='cubic', bounds_error=False)
    data_interpolated[column] = interpolator(database.index)
```

Besides the data for indices, I decided to calculate and data for moving averages and technical indicators. This decision comes with the necesity to better understand the market fluctuation and volatility, and to capture main components.

```python
# 1. Moving Averages
data_interpolated['BTC_7D_MA'] = data_interpolated['BTC_Close'].rolling(window=7).mean()
data_interpolated['BTC_30D_MA'] = data_interpolated['BTC_Close'].rolling(window=30).mean()

# 2. Technical Indicators
# Example: Relative Strength Index (RSI)
def calculate_rsi(df, window=14):
    delta = data_interpolated['BTC_Close'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()
    rs = gain / loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

data_interpolated['RSI'] = calculate_rsi(data_interpolated)

# 3. Moving Average Convergence Divergence (MACD)
def calculate_macd(data_interpolated, short_window=12, long_window=26, signal_window=9):
    short_ema = data_interpolated['BTC_Close'].ewm(span=short_window, min_periods=1, adjust=False).mean()
    long_ema = data_interpolated['BTC_Close'].ewm(span=long_window, min_periods=1, adjust=False).mean()
    macd = short_ema - long_ema
    signal = macd.ewm(span=signal_window, min_periods=1, adjust=False).mean()
    return macd, signal

data_interpolated['MACD'], data_interpolated['MACD_Signal'] = calculate_macd(data_interpolated)

# 4. Volatility Measures
# Example: Historical Volatility (30-day)
data_interpolated['BTC_30D_volatility'] = data_interpolated['BTC_Close'].pct_change().rolling(window=30).std() * (252 ** 0.5)  # 252 trading days in a year

# 5. Volume-based Features
data_interpolated['BTC_7D_Volume_MA'] = data_interpolated['BTC_Volume'].rolling(window=7).mean()

# 6. Calendar Features
data_interpolated['Year'] = data_interpolated['Date'].dt.year
data_interpolated['Month'] = data_interpolated['Date'].dt.month
data_interpolated['DayOfWeek'] = data_interpolated['Date'].dt.dayofweek

# Drop rows with NaN values resulting from moving averages or rolling calculations
data_interpolated.dropna(inplace=True)

data_interpolated.head(10)
```

Splitiing the data

The data willbe splitted by date. If we are splitting the data using another method, like 'train_test_split' function, the data will be splitte randmomly and this will not be helpful for our forecasts.

```python
from datetime import timedelta

X = ['BTC_Close','DJ_Close', 'SP_Close', 'NDQ_Close', 'BTC_7D_MA',
       'BTC_30D_MA', 'RSI', 'MACD', 'MACD_Signal', 'BTC_30D_volatility',
       'BTC_7D_Volume_MA', 'Year', 'Month', 'DayOfWeek']

# Find the last month date
last_month_date = max(data_interpolated['Date']) - timedelta(days=30)

# Create boolean masks for filtering
train_mask = data_interpolated['Date'] <= last_month_date
test_mask = data_interpolated['Date'] > last_month_date

# Filter the data for training and testing
X_train = data_interpolated.loc[train_mask, X]
X_test = data_interpolated.loc[test_mask, X]
y_train = data_interpolated.loc[train_mask, 'BTC_Shifted']
y_test = data_interpolated.loc[test_mask, 'BTC_Shifted']
```

Training the models

We start by training some classical algorithms and then move to Neural Networks and Deep Learning algorithms. Training multiple algorithms, will offer a better view when choosing the appropriate model.

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, ExtraTreesRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Define models
models = {
    "Random Forest": RandomForestRegressor(),
    "Extra Trees": ExtraTreesRegressor(),
    "Gradient Boosting": GradientBoostingRegressor(),
    "KNeighbours": KNeighborsRegressor(),
    "SVR": SVR(),
    "XGBRegressor": XGBRegressor()
}

# Dictionary to store MAE values for each model
mae_values_classical = {}

# Fit models and evaluate
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mae = mean_absolute_error(y_test, y_pred)
    mae_values_classical[name] = mae  # Store MAE value for the current model
    
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    print(name + ":")
    print("MAE:", mae)
    print("MSE:", mse)
    print("R^2 Score:", r2)
    print()
```

The ARIMA and SARIMAX models will need a good parameter order to work correctly. Since we are not using auto-arima, we need to do some statistical anlaysis to choose the appropriate parameter order. 

```python
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Perform the Augmented Dickey-Fuller test for stationarity
adf_result = adfuller(data_interpolated['BTC_Shifted'].dropna())
print('ADF Statistic:', adf_result[0])
print('p-value:', adf_result[1])
print('Critical Values:')
for key, value in adf_result[4].items():
    print(f'\t{key}: {value}')

# Plot autocorrelation and partial autocorrelation plots
fig, ax = plt.subplots(2, 1, figsize=(10, 8))
plot_acf(data_interpolated['BTC_Shifted'].dropna(), ax=ax[0], lags=50)
plot_pacf(data_interpolated['BTC_Shifted'].dropna(), ax=ax[1], lags=50)
plt.show()
```

```python
from statsmodels.tsa.stattools import kpss

# Perform KPSS test
kpss_stat, p_value, lags, critical_values = kpss(data_interpolated['BTC_Shifted'])

print('KPSS Statistic:', kpss_stat)
print('p-value:', p_value)
print('Lags Used:', lags)
print('Critical Values:')
for key, value in critical_values.items():
    print(f'\t{key}: {value}')
```

```python
# Model evaluation

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Fit ARIMA models
model_arima_110 = ARIMA(y_train, order=(1, 1, 0)).fit()
model_arima_111 = ARIMA(y_train, order=(1, 1, 1)).fit()

# Residual Analysis
def residual_analysis(model, name):
    # Calculate residuals
    residuals = model.resid

    # Plot residuals over time
    plt.figure(figsize=(10, 6))
    plt.plot(residuals)
    plt.title(f'{name} Residuals')
    plt.xlabel('Time')
    plt.ylabel('Residuals')
    plt.show()

    # Check for autocorrelation in residuals
    # Perform Ljung-Box test
    ljung_box_test = model.test_serial_correlation(method='ljungbox')
    
    if isinstance(ljung_box_test, tuple):
        p_values = ljung_box_test[1]
        print(f'Ljung-Box Test (p-values): {p_values}')
    else:
        print(f'Ljung-Box Test (p-value): {ljung_box_test}')

# Evaluate Out-of-Sample Forecasting
def evaluate_forecasting(model, name):
    # Make forecasts for the testing dataset
    forecasts = model.forecast(steps=len(y_test))

    # Calculate evaluation metrics
    mae = mean_absolute_error(y_test, forecasts)
    mse = mean_squared_error(y_test, forecasts)
    rmse = np.sqrt(mse)
    mape = np.mean(np.abs((y_test - forecasts) / y_test)) * 100

    print(f'{name} Evaluation Metrics:')
    print(f'Mean Absolute Error (MAE): {mae}')
    print(f'Mean Squared Error (MSE): {mse}')
    print(f'Root Mean Squared Error (RMSE): {rmse}')
    print(f'Mean Absolute Percentage Error (MAPE): {mape}%')

    return mae

# Information Criteria
def information_criteria(model, name):
    aic = model.aic
    bic = model.bic

    print(f'{name} Information Criteria:')
    print(f'Akaike Information Criterion (AIC): {aic}')
    print(f'Bayesian Information Criterion (BIC): {bic}')

# Visual Inspection
def visual_inspection(model, name):
    # Plot observed vs. predicted values
    plt.figure(figsize=(10, 6))
    plt.plot(y_test, label='Observed')
    plt.plot(model.forecast(steps=len(y_test)), label='Predicted')
    plt.title(f'{name} Forecasting')
    plt.xlabel('Time')
    plt.ylabel('BTC Price')
    plt.legend()
    plt.show()

# Residual Analysis
residual_analysis(model_arima_110, 'ARIMA(1,1,0)')
residual_analysis(model_arima_111, 'ARIMA(1,1,1)')

# Evaluate Out-of-Sample Forecasting
evaluate_forecasting(model_arima_110, 'ARIMA(1,1,0)')
evaluate_forecasting(model_arima_111, 'ARIMA(1,1,1)')

# Information Criteria
information_criteria(model_arima_110, 'ARIMA(1,1,0)')
information_criteria(model_arima_111, 'ARIMA(1,1,1)')

# Visual Inspection
visual_inspection(model_arima_110, 'ARIMA(1,1,0)')
visual_inspection(model_arima_111, 'ARIMA(1,1,1)')

mae_arima_110 = evaluate_forecasting(model_arima_110, 'ARIMA(1,1,0)')
mae_arima_111 = evaluate_forecasting(model_arima_111, 'ARIMA(1,1,1)')
```

The last model we are going to use is LSTM. For this model the DataFrames must be reshaped to fit correctly. We are also defining another variables to avoid confusion.

```python
from datetime import timedelta

# Define the columns to keep
columns_to_keep = ['BTC_Close', 'DJ_Close', 'SP_Close', 'NDQ_Close', 'BTC_Shifted', 'BTC_7D_MA',
                   'BTC_30D_MA', 'RSI', 'MACD', 'MACD_Signal', 'BTC_30D_volatility',
                   'BTC_7D_Volume_MA', 'Year', 'Month', 'DayOfWeek']

# Extract the relevant columns
data_selected = data_interpolated[columns_to_keep]

# Find the last month date
last_month_date = max(data_interpolated['Date']) - timedelta(days=30)

# Create boolean masks for filtering
train_mask1 = data_interpolated['Date'] <= last_month_date
test_mask1= data_interpolated['Date'] > last_month_date

# Filter the data for training and testing
X_train1 = data_selected[train_mask1]
X_test1 = data_selected[test_mask1]
y_train1 = data_interpolated.loc[train_mask1, 'BTC_Shifted']
y_test1 = data_interpolated.loc[test_mask1, 'BTC_Shifted']
```

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout
from sklearn.preprocessing import MinMaxScaler

# Normalize the data
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train1)
X_test_scaled = scaler.transform(X_test1)

# Reshape the data for LSTM input [samples, time steps, features]
n_features = len(columns_to_keep)
time_steps = 1  # You can adjust this based on your preference
X_train_reshaped = X_train_scaled.reshape((X_train_scaled.shape[0], time_steps, n_features))
X_test_reshaped = X_test_scaled.reshape((X_test_scaled.shape[0], time_steps, n_features))

# Define the LSTM model
model = Sequential()

# Add the first LSTM layer
model.add(LSTM(units=50, activation='relu', return_sequences=True, input_shape=(time_steps, n_features)))
model.add(Dropout(0.2))

# Add the second LSTM layer
model.add(LSTM(units=50, activation='relu', return_sequences=True))
model.add(Dropout(0.2))

# Add the third LSTM layer
model.add(LSTM(units=50, activation='relu', return_sequences=False))  # No need to return sequences for the last layer
model.add(Dropout(0.2))

# Add the output layer
model.add(Dense(units=1))

# Compile the model
model.compile(optimizer='adam', loss='mean_absolute_error')

# Fit the model
history = model.fit(X_train_reshaped, y_train1, epochs=100, batch_size=32, validation_data=(X_test_reshaped, y_test1), verbose=1)
```

```python
import numpy as np
from sklearn.metrics import mean_absolute_error

# Make predictions
y_train_pred = model.predict(X_train_reshaped)
y_test_pred = model.predict(X_test_reshaped)

# Calculate MAE
train_mae = mean_absolute_error(y_train1, y_train_pred)
test_mae = mean_absolute_error(y_test1, y_test_pred)
mae_LSTM = test_mae
print("Train MAE:", train_mae)
print("Test MAE:", test_mae)


# You can also visualize the predictions vs actuals
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(y_test.index, y_test.values, label='Actual')
plt.plot(y_test.index, y_test_pred, label='Predicted')
plt.title('BTC Price Prediction')
plt.xlabel('Date')
plt.ylabel('BTC Price')
plt.legend()
plt.show()
```
